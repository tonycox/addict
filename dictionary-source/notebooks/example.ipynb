```python
from llama_cpp import Llama
import json
import multiprocessing
```

```python
max_tokens = 2048
n_batch = 2048
temperature = 0.6
verbose = True
f16_kv = True
```

```python
# GPT3.5 - 1.3B, 6B, and 175B parameters
# current Llama - 13B parameters

model = Llama(
    model_path=model_path,
    n_ctx=max_tokens,
    n_batch=n_batch,
    n_threads=multiprocessing.cpu_count(),
    use_mlock=True,
    f16_kv=f16_kv,
    verbose=verbose,
)
```
